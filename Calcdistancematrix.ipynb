```python
multiprocess = pd.read_csv('C:/Users/Admin_2/Documents/Thesis/NewExperiments/test_multiprocessnew.csv')
multiprocess 
```

```python
#1. Load the kappa file two times
mp1 = pd.read_csv('C:/Users/Admin_2/Documents/Thesis/NewExperiments/test_multiprocessnew.csv')
mp2 = pd.read_csv('C:/Users/Admin_2/Documents/Thesis/NewExperiments/test_multiprocessnew.csv')
```

```python
#switch map 1 with map 2 for mp2
mp2 = mp2[["Unnamed: 0", "map2", "map1", 'kappa']]
mp2.columns = ["Unnamed: 0", "map1", "map2", 'kappa']
mp2
```

```python
#2. Merge the two data frames together
frames = [mp1, mp2]
mpwhole = pd.concat(frames)
mpwhole
```

```python
#4. Make sure the list has same order as experiments
listnames = mpwhole.map1.unique()
strs = listnames
listend = strs[:1999]
listend
listbegin = strs[1999:2000]
listbegin
wholelist = np.append(listbegin, listend)
wholelist
```

Next, an empty data frame is created with the "wholelist" as index and columns:

```python
dfempty = pd.DataFrame(index=wholelist, columns=wholelist)
dfempty
```

```python
list(dfempty.columns)
```

Also, a distance matrix is created from the merged datafram df4 to get kappa_df. 
However, this kappa_df has not yet the same order as the experiments, which will be fixed later using the empty dataframe.

```python
kappa_df = mpwhole.pivot(index='map2', columns='map1', values='kappa')
kappa_df
```

```python
#Fill the NAN values with value 1, which means these map pairs are the same
kappa_df = kappa_df.fillna(1)
kappa_df
```

```python
len(kappa_df)
```

Using the kappa_df and the empty dataframe, with "reindex" a kappa dataframe is created which has the same order as the experiments:

```python
kappa_df = kappa_df.reindex(dfempty.index)
```

```python
kappa_df=kappa_df.T
kappa_df=kappa_df.reindex(dfempty.index)
kappa_df
```

Safe kappa file as csv:
